```python
def make_ts(i):
    if i%5==0:
        return []
    else:
        n=np.random.randint(1,100)
        R=0.1
        n1=np.random.binomial(n, R)
        n2=n-n1
        ts=np.concatenate((np.random.exponential(5, size=n1), np.random.exponential(30, size=n2)))
        ind=np.nonzero(np.random.choice(2, size=len(ts), p=[0.7, 0.3])==1)[0]
        ts=ts[ind]
    return ts
```

```python
events=[]
for i in range(Nevents):
    print('Event', i)
    ts=make_ts(i)
    event=Event('sim', SPE, area_sigma, pmt)
    event.times=np.sort(ts)
    noise=Noise[np.random.randint(0,np.shape(Noise)[0]), pmt]   
    event.make_wf(noise)    
    events.append(event)

file = open(path+t_or_v+'_sim_events/PMT{}'.format(pmt), "wb")
pkl.dump(events, file)
file.close()
```
